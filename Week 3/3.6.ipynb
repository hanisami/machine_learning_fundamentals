#### Start by generating the data for our linear regression example:

```python
import numpy as np

# Generate random data for x
np.random.seed(42)  # for reproducibility
x = np.random.rand(100, 1)

# Generate corresponding y values using the equation y = 3x + 4 + noise
y = (3 + 4 * x) + np.random.randn(100, 1) * 0.5
```

```python
import matplotlib.pyplot as plt

# Plot the scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, label='Data Points')

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot of Generated Data')

# Show the plot
plt.legend()
plt.grid(True)
plt.show()
```

#### Now that we have our data, let's implement linear regression step by step:

1. Initialize Parameters: We start by initializing the parameters of our model, which are the slope (a) and the intercept (b). Let's set them to random values to begin with.

```python
# Initialize parameters
np.random.seed(42)
a = np.random.randn(1)
b = np.random.randn(1)
```

2. Compute Predictions: Next, we compute the predictions of our model using the current parameters.

```python
# Compute predictions
predictions = a * x + b
print(predictions)
```

3. Compute Loss: Now, we compute the Mean Squared Error (MSE) loss between our predictions and the actual values.

```python
# Compute loss
loss = np.mean((predictions - y) ** 2)
```

4. Compute Gradients: We compute the gradients of the loss function with respect to each parameter using the derivative of the loss function.

```python
# Compute gradients
grad_a = np.mean(2 * (predictions - y) * x)
grad_b = np.mean(2 * (predictions - y))
```

5. Update Parameters: Finally, we update the parameters of our model using Gradient Descent.

```python
# Considering learning rate value of 0.01
# Update parameters using Gradient Descent
learning_rate = 0.01
a -= learning_rate * grad_a
b -= learning_rate * grad_b
```

6. Repeat: We repeat steps 2-5 until convergence, i.e., until the loss stops decreasing or after a fixed number of iterations.

```python
# Number of iterations
num_iterations = 1000

# Learning rate
learning_rate = 0.01

# Gradient Descent
for i in range(num_iterations):
    # Compute predictions
    predictions = a * x + b
    
    # Compute loss
    loss = np.mean((predictions - y) ** 2)
    
    # Compute gradients
    grad_a = np.mean(2 * (predictions - y) * x)
    grad_b = np.mean(2 * (predictions - y))
    
    # Update parameters
    a -= learning_rate * grad_a
    b -= learning_rate * grad_b
    
    # Print progress
    if i % 100 == 0:
        print(f'Iteration {i}, Loss: {loss}')

print(f'Final Parameters: slope = {a}, intercept = {b}')
```

```python
import matplotlib.pyplot as plt

# Plot the scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, label='Data Points')

# Compute predictions using updated parameters
predictions = a * x + b

# Plot the resulting line
plt.plot(x, predictions, color='red', label='Regression Line')

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression with Gradient Descent')

# Show legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
# Assuming the linear regression model has been trained and parameters 'a' and 'b' have been obtained

# Generate testing data
x_test = np.random.rand(50, 1)
y_test = (3 + 4 * x_test) + np.random.randn(50, 1) * 0.5

# Compute predictions on testing data
predictions_test = a * x_test + b

# Compute loss on testing data (Mean Squared Error)
loss_test = np.mean((predictions_test - y_test) ** 2)

# Compute loss on training data (if not already computed)
predictions_train = a * x + b
loss_train = np.mean((predictions_train - y) ** 2)

# Print the loss values
print(f'Training Loss: {loss_train}')
print(f'Testing Loss: {loss_test}')
```

```python
from sklearn.model_selection import learning_curve
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np

# Generate data
x = np.random.rand(100, 1)
y = (3 + 4 * x) + np.random.randn(100, 1) * 0.5

# Initialize linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(x, y)

# Plot learning curve
train_sizes, train_scores, test_scores = learning_curve(estimator=model, X=x, y=y, train_sizes=np.linspace(0.1, 1.0, 10), cv=5, scoring='neg_mean_squared_error')

train_scores_mean = np.mean(-train_scores, axis=1)
test_scores_mean = np.mean(-test_scores, axis=1)

plt.figure(figsize=(10, 6))
plt.plot(train_sizes, train_scores_mean, label='Training error')
plt.plot(train_sizes, test_scores_mean, label='Validation error')
plt.xlabel('Training examples')
plt.ylabel('Negative Mean Squared Error')
plt.title('Learning Curve')
plt.legend()
plt.grid()
plt.show()
```
