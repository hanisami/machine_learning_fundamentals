```python
import numpy as np
x = np.arange(0,100).reshape(100,1)
y = (3 + (4 * x)) + np.random.randint(-50, 100, (100, 1))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
ax = sns.scatterplot(x.flatten(),y.flatten())
ax.set(xlabel='X', ylabel='Y')
plt.show()
```

```python
# y = 3 * x + 4
```

```python
# lr = [3*x_i + 3 for x_i in x]
lr1 = [4*x_i + 40 for x_i in x]
lr2 = [5*x_i + 80 for x_i in x]
lr3 = [6*x_i + 120 for x_i in x]
lr4 = [7*x_i + 160 for x_i in x]
lr5 = [2*x_i + 10 for x_i in x]
```

```python
# Plot scatter plot of x and y
sns.scatterplot(x=x.flatten(), y=y.flatten())

# Plot linear regression line
sns.lineplot(x=x.flatten(), y=np.array(lr1).flatten(), color='red', label='y = 4x + 40')
sns.lineplot(x=x.flatten(), y=np.array(lr2).flatten(), color='green', label='y = 5x + 80')
sns.lineplot(x=x.flatten(), y=np.array(lr3).flatten(), color='black', label='y = 6x + 120')
sns.lineplot(x=x.flatten(), y=np.array(lr4).flatten(), color='orange', label='y = 7x + 160')
sns.lineplot(x=x.flatten(), y=np.array(lr5).flatten(), color='purple', label='y = 2x + 10')



# Set plot labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression')

# Show the plot
plt.show()
```
