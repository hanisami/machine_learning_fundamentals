### Data Preprocessing

```python
# we will use KNN algorthim so we must first preprocess our data so let's start
from sklearn import preprocessing  
label_encoder = preprocessing.LabelEncoder()  
df['diagnosis']= label_encoder.fit_transform(df['diagnosis']) 
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
scale = scaler.fit_transform(df) 
print(scale)
```

### Split the data into train and test sets

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop(columns=['diagnosis','id'],axis=1)
y=df['diagnosis']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42)
```

### Build and train the KNeighborsClassifier

- Train the classifier on different values for n_neighbors (from 1 to 11) using the 'Grid Search' and compare between the values
- The evaluation should be on both, the training and test sets.

```python
from sklearn.neighbors import KNeighborsClassifier 
```

```python
training_accuracy=[]
testing_accuracy=[]
n=range(1,11)
for i in range(1,11):
    clf=KNeighborsClassifier(n_neighbors=i)
    clf.fit(X_train,y_train)
    testing_accuracy.append(clf.score(X_test,y_test))
    training_accuracy.append(clf.score(X_train,y_train))
    
for i in range(0,10):
    print("when the number of  k was "+str(i+1))
    print("the accuracy of testing was "+str(testing_accuracy[i]))
    print("the accuracy of trainning was "+ str(training_accuracy[i]))
```

### Plot the training and test accuracy

- Use matplotlib to draw a single line chart showing the change in accuracy with respect to the number of neighbors

```python
import matplotlib.pyplot as plt
plt.plot(n,training_accuracy,label="training accuracy")
plt.plot(n,testing_accuracy,label="testing_accuracy")
plt.ylabel("accuracy")
plt.xlabel("number of K ")
plt.legend()
```
