# Decision Tree Classification Application

Decision Tree algorithm in brief

Decision tree is a supervised machine learning algorithm mainly used for Regression and Classification. While dividing a dataset into smaller and smaller subsets, a corresponding decision tree is gradually developed. The final result is a tree with decision nodes and leaf nodes.

Source : https://www.geeksforgeeks.org/decision-tree/

- It forms the basis of random forest. The aim is to transform complex structures in the data set into simple decision structures. Heterogeneous data sets are divided into homogeneous subgroups according to a specified target variable. The machine finds the best decision rule.
- With SSE for regression problems
- We use Gini and Entropy in classification problems.
Since this is related to the data set classification problem, I will talk about gini and entropy.


Gini Impurity is a score that evaluates how accurate a split is among the classified groups. The Gini Impurity evaluates a score in the range between 0 and 1, where 0 is when all observations belong to one class, and 1 is a random distribution of the elements within classes. In this case, we want to have a Gini index score as low as possible. Gini Index is the evaluation metric we shall use to evaluate our Decision Tree Model.

ENTROPY

When the data is divided according to a feature, the best choice is made if the uncertainty of each data set is minimum and therefore the information gain is maximum.Entropy represents randomness, uncertainty, and the probability of the unexpected occurring. This distinctive feature is achieved by calculating the information gain or gini index. If all the samples are regular/homogeneous, their entropy will be zero These operations are calculated separately for each feature.

- These measurements are called purity measurements.
- It expresses the relationship between similarities and dissimilarities. It provides the opportunity to evaluate the error in entopy and Gini classification problems.
- Gini index is a metric that measures the homogeneity of a data set. If the Gini index is high, this indicates that the data set has a non-homogeneous, complex or mixed distribution.
- If the entropy is high, this means that the dataset has a more complex and inhomogeneous distribution. That is, samples of different classes are mixed together and it is more difficult to form a homogeneous subset

```python
X = df.drop(['class'], axis=1)
y = df['class']
```

```python
# split X and y into training and test sets
from sklearn.model_selection import train_test_split 
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33, random_state=42)
```

```python
# checking the shape of train and test 
X_train.shape,X_test.shape 
```

# Future Engineering

```python
X_train.dtypes
```

```python
X_train.head()
```

```python
# encoding process in categorical columns
import category_encoders as ce
```

```python
# encode categorical variables with ordinal encoding

encoder = ce.OrdinalEncoder(cols=['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety'])

X_train = encoder.fit_transform(X_train)

X_test = encoder.transform(X_test)
```

```python
X_train.head()
```

```python
X_test.head()
```

# Decision Tree Classifier 

```python
# import DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc_gini = DecisionTreeClassifier(criterion='gini', max_depth=3, min_samples_split= 2, random_state=0, min_samples_leaf= 1)


# fit the model
dtc_gini.fit(X_train, y_train)
```

```python
y_pred_gini = dtc_gini.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print('Model accuracy score with criterion gini index: {0:0.4f}'. format(accuracy_score(y_test, y_pred_gini)))
```

### Compare the train-set and test-set accuracy

Now, I will compare the train-set and test-set accuracy to check for overfitting.

```python
y_pred_train_gini = dtc_gini.predict(X_train)

y_pred_train_gini
```

```python
print('Training-set accuracy score: {0:0.4f}'. format(accuracy_score(y_train, y_pred_train_gini)))
```

```python
# print the scores on training and test set

print('Training set score: {:.4f}'.format(dtc_gini.score(X_train, y_train)))

print('Test set score: {:.4f}'.format(dtc_gini.score(X_test, y_test)))
```

```python
plt.figure(figsize=(12,8))

from sklearn import tree

tree.plot_tree(dtc_gini.fit(X_train, y_train)) 
```

```python
import graphviz 
dot_data = tree.export_graphviz(dtc_gini, out_file=None, 
                              feature_names=X_train.columns,  
                              class_names=y_train,  
                              filled=True, rounded=True,  
                              special_characters=True)

graph = graphviz.Source(dot_data) 

graph 
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred_gini))
```

```python
# instantiate the DecisionTreeClassifier model with criterion entropy

dtc_en = DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=0, min_samples_leaf= 1)


# fit the model
dtc_en.fit(X_train, y_train)
```

```python
y_pred_en = dtc_en.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print('Model accuracy score with criterion entropy: {0:0.4f}'. format(accuracy_score(y_test, y_pred_en)))
```

```python
y_pred_train_en = dtc_en.predict(X_train)

y_pred_train_en
```

```python
print('Training-set accuracy score: {0:0.4f}'. format(accuracy_score(y_train, y_pred_train_en)))
```

```python
# print the scores on training and test set

print('Training set score: {:.4f}'.format(dtc_en.score(X_train, y_train)))

print('Test set score: {:.4f}'.format(dtc_en.score(X_test, y_test)))
```

```python
plt.figure(figsize=(12,8))

from sklearn import tree

tree.plot_tree(dtc_en.fit(X_train, y_train))
```

```python
import graphviz 
dot_data = tree.export_graphviz(dtc_en, out_file=None, 
                              feature_names=X_train.columns,  
                              class_names=y_train,  
                              filled=True, rounded=True,  
                              special_characters=True)

graph = graphviz.Source(dot_data) 

graph
```
