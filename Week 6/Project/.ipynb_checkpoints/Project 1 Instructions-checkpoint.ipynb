### Clustering 

Using KMeans and Silhoutte scores

To find the best number of cluster (n_clusters=k) we run a for-loop and compute the WSS (Within sum of squares) Elbow method and Silhoutte scores for each "k".

**BMI**

```python
wss=[]
sil=[]
for k in range(2,16):
    kmeans=KMeans(n_clusters=k, random_state=1).fit(bmi_charg_c)
    wss.append(kmeans.inertia_)
    labels=kmeans.labels_
    silhoutte=silhouette_score(bmi_charg_c, labels, metric = 'euclidean')
    sil.append(silhoutte)
```

```python
k=range(2,16)
style.use("bmh")
fig,ax=plt.subplots(figsize=(14,6))
ax.set_facecolor("white")
ax.plot(k, wss, color="green")
ax.xaxis.set_major_locator(MaxNLocator(nbins=15, integer=True))
ax.set_xlabel("No of clusters", fontsize=20)
ax.set_ylabel("WSS (With in Sum of squares)", fontsize=20)
ax2=ax.twinx()
ax2.plot(k, sil, color="blue")
ax2.set_ylabel("Silhouette scores", fontsize=20)
ax2.grid(True,color="silver")
plt.show()
```

From the plot we see the "elbow" at 3 and silhouutee score almost best at that point.

```python
k=3
kmeans=KMeans(n_clusters=k, random_state=1).fit(bmi_charg_c)
clusters=kmeans.labels_
centrids=kmeans.cluster_centers_
raw_data["clusters"]=clusters
raw_data
```

```python
raw_data2=raw_data.sort_values(["clusters"]).copy()
```

```python
for i in range(0,k+1):
    raw_data2["clusters"]=raw_data2["clusters"].replace(i, chr(i+65))
    
raw_data2
```

## DBSCAN Method

```python
from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
import numpy as np

# Assuming you have imported raw_data2 and applied preprocessing if needed

# Extract features for clustering
x=raw_data2.iloc[:,[0,6]].values

# Step 1: Use the elbow method to determine epsilon
# Calculate the distance to the k=4 nearest neighbor for each point
k = 4
nbrs = NearestNeighbors(n_neighbors=k).fit(x)
distances, _ = nbrs.kneighbors(x)
distances = np.sort(distances, axis=0)
distances = distances[:,1]  # Take the distance to the k=4 nearest neighbor (not including the point itself)

# Plot the distances
plt.plot(distances)
plt.xlabel('Samples')
plt.ylabel('Distance to 4th Nearest Neighbor')
plt.title('Elbow Method for Optimal Epsilon (k=4)')
plt.show()

# From the plot, determine a suitable epsilon value
epsilon = 170  # Example value, adjust as needed

# Step 2: Choose min_samples
min_samples = 3  # Example value, adjust as needed

# Step 3: Perform DBSCAN clustering with tuned parameters
dbscan = DBSCAN(eps=epsilon, min_samples=min_samples)
y_dbscan = dbscan.fit_predict(x)

# Visualize the clustering results (if desired)
# (Example visualization code)
plt.scatter(x[:, 0], x[:, 1], c=y_dbscan, cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('DBSCAN Clustering with Tuned Parameters')
plt.colorbar(label='Cluster Label')
plt.show()
```

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score

# Define a range of min_samples values to explore
min_samples_range = range(2, 21)

# Initialize variables to store results
best_min_samples = None
best_silhouette_score = -1  # Initialize with a low value

# Iterate over min_samples values
for min_samples in min_samples_range:
    # Perform DBSCAN clustering
    dbscan = DBSCAN(min_samples=min_samples)
    cluster_labels = dbscan.fit_predict(x)
    
    # Check if more than one cluster is formed
    unique_labels = len(set(cluster_labels))
    if unique_labels > 1:
        # Compute silhouette score
        silhouette_avg = silhouette_score(x, cluster_labels)
        
        # Update best parameters if silhouette score is improved
        if silhouette_avg > best_silhouette_score:
            best_silhouette_score = silhouette_avg
            best_min_samples = min_samples

print("Best min_samples:", best_min_samples)
print("Best silhouette score:", best_silhouette_score)
```

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score

# Define a range of eps values to explore
eps_range = [0.1, 0.5, 1.0, 1.5, 2.0]  # Example range, adjust as needed

# Initialize variables to store results
best_eps = None
best_silhouette_score = -1  # Initialize with a low value

# Iterate over eps values
for eps in eps_range:
    # Perform DBSCAN clustering
    dbscan = DBSCAN(eps=eps, min_samples=5)  # Example min_samples, adjust as needed
    cluster_labels = dbscan.fit_predict(x)
    
    # Check if more than one cluster is formed
    unique_labels = len(set(cluster_labels))
    if unique_labels > 1:
        # Compute silhouette score
        silhouette_avg = silhouette_score(x, cluster_labels)
        
        # Update best parameters if silhouette score is improved
        if silhouette_avg > best_silhouette_score:
            best_silhouette_score = silhouette_avg
            best_eps = eps

print("Best eps:", best_eps)
print("Best silhouette score:", best_silhouette_score)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Assuming x contains your data points

# Initialize an empty list to store silhouette scores
silhouette_scores = []

# Specify the range of cluster numbers you want to try
min_clusters = 2
max_clusters = 10

# Iterate over different numbers of clusters
for n_clusters in range(min_clusters, max_clusters + 1):
    # Perform KMeans clustering with the current number of clusters
    kmeans = KMeans(n_clusters=n_clusters)
    cluster_labels = kmeans.fit_predict(x)
    
    # Compute the silhouette score for this clustering
    silhouette_avg = silhouette_score(x, cluster_labels)
    
    # Append the silhouette score to the list
    silhouette_scores.append(silhouette_avg)

# Plot the silhouette scores
plt.figure(figsize=(10, 6))
plt.plot(range(min_clusters, max_clusters + 1), silhouette_scores, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score vs. Number of Clusters')
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

# Assuming x contains your data points
# Assuming best_n_clusters is the optimal number of clusters determined from the silhouette score

# Perform Agglomerative Clustering with the optimal number of clusters
agg_clustering = AgglomerativeClustering(n_clusters=3)
cluster_labels = agg_clustering.fit_predict(x)

# Plot the clustering results
plt.figure(figsize=(10, 6))
plt.scatter(x[:, 0], x[:, 1], c=cluster_labels, cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title(f'Agglomerative Clustering with {3} Clusters')
plt.colorbar(label='Cluster Label')
plt.show()
```
